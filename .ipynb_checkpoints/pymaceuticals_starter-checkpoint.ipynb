```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
study_complete = study_results.merge(mouse_metadata, left_on = "Mouse ID", right_on = "Mouse ID")

# Display the data table for preview

# note: it looks like my merge worked differently, so my row indices are different
study_complete.head()
```

```python
# Checking the number of mice.
len(study_complete['Mouse ID'].unique())
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.

duplicate_id = study_complete[study_complete.duplicated(subset = ['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
duplicate_id
```

```python
duplicate_data = study_complete.loc[study_complete['Mouse ID'] == 'g989']
duplicate_data
```

```python
study_clean = study_complete[study_complete['Mouse ID'] != 'g989']
study_clean
```

```python
# Checking the number of mice in the clean DataFrame.
len(study_clean['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean_tumor_vol = study_clean.groupby("Drug Regimen").mean()['Tumor Volume (mm3)']
med_tumor_vol = study_clean.groupby("Drug Regimen").median()['Tumor Volume (mm3)']
var_tumor_vol = study_clean.groupby("Drug Regimen").var()['Tumor Volume (mm3)']
std_tumor_vol = study_clean.groupby("Drug Regimen").std()['Tumor Volume (mm3)']
sem_tumor_vol = study_clean.groupby("Drug Regimen").sem()['Tumor Volume (mm3)']


regimen_grouped = pd.DataFrame({
    'Mean Tumor Volume' : mean_tumor_vol,
    'Median Tumor Volume': med_tumor_vol,
    'Tumor Volume Variance': var_tumor_vol,
    'Tumor Volume Std. Dev.': std_tumor_vol,
    'Tumor Volume Std. Err.': sem_tumor_vol
})

regimen_grouped
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line

regimen_grouped = study_clean.groupby("Drug Regimen").agg({
    'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']
})

regimen_grouped
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# subsetting for four treatment regimens
regimen_subset = study_clean[study_clean['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Start by getting the last (greatest) timepoint for each mouse
regimen_subset = regimen_subset.groupby(['Mouse ID']).last().reset_index()
#regimen_subset = regimen_subset.drop_duplicates(subset = ['Mouse ID'], keep = 'last')

# taking only mouse ID and last tumor volume
regimen_subset = regimen_subset[['Mouse ID', 'Drug Regimen', 'Tumor Volume (mm3)', 'Timepoint']]

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
#final_tumor_volume = study_clean.merge(regimen_subset, left_on = 'Mouse ID', right_on = 'Mouse ID', how = 'right')
final_tumor_volume = regimen_subset.rename(columns = {'Tumor Volume (mm3)': 'Final Tumor Volume',
                                                          'Timepoint': 'Final Timepoint'})
final_tumor_volume.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
regimens = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []
```

```python

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for regimen in regimens:
    
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    volumes = final_tumor_volume.loc[final_tumor_volume["Drug Regimen"] == regimen]['Final Tumor Volume']
    
    # adding to list
    tumor_vol_data.append(list(volumes))
    
    # creating quantiles for final tumor volume of each drug regimen
    quartiles = volumes.quantile([0.25, 0.5, 0.75])
    
    # taking lower and upper quantiles
    lower_q = quartiles[0.25]
    upper_q = quartiles[0.75]
    
    # calculating iqr for for final tumor volume for given drug regimen
    iqr = upper_q - lower_q
    
    # calculating boundaries for outliers
    lower = lower_q - (1.5 * iqr)
    upper = upper_q + (1.5 * iqr)
    
    # creating output message
    print(regimen + ' IQR: ' + str(iqr))
    
    message = regimen + "\'s potential outliers: "
    message = message + str(volumes.loc[(volumes > upper) | (volumes < lower)]) + '\n'
    print(message)
    # Determine outliers using upper and lower bounds
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

plt.boxplot(tumor_vol_data, labels = regimens, sym = 'r')
plt.ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin

# filtering for mouse l509
one_mouse = study_clean[study_clean['Mouse ID'] == 'l509']
plt.plot(one_mouse['Timepoint'], one_mouse['Tumor Volume (mm3)'])

#adding axis labels and titles
plt.xlabel('Timepoint (days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Capomulin treatment of mouse l509')
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

# filtering for capomulin subjects
capomulin_regimen = study_clean[study_clean['Drug Regimen'] == 'Capomulin']
# grouping by average tumor volume for each mouse
capomulin_regimen = capomulin_regimen.groupby('Mouse ID').mean()

# plotting weight vs tumor volume
plt.scatter(x = 'Weight (g)', y = 'Tumor Volume (mm3)', data = capomulin_regimen)
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

# calculating values for linear regression model
slope, intercept, r, p, std_err = st.linregress(capomulin_regimen['Weight (g)'], capomulin_regimen['Tumor Volume (mm3)'])
# displaying pearson's r coefficient
print("The correlation between mouse weight and the average tumor volume is:", round(r, 2))
```

```python
# calculating points for line of best fit
fit = (slope * capomulin_regimen['Weight (g)']) + intercept

# plotting weight vs tumor volume
plt.scatter(x = 'Weight (g)', y = 'Tumor Volume (mm3)', data = capomulin_regimen)
# plotting line of best fit
plt.plot(capomulin_regimen['Weight (g)'], fit, color = 'r')
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```
